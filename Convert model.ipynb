```python
import tensorflow as tf

# Load the .keras model
model = tf.keras.models.load_model("/tmp/inception_v3_Adam_unfreeze_10_epoch_10.keras")
```

```python
model.layers
```

```python
model_without_preprocessing = tf.keras.Sequential([
    model.layers[0],  # Input
    model.layers[4],  # Inception
    model.layers[6],  # Flatten
    model.layers[7],   # Dense 1
    model.layers[8],  # Dense 2
    model.layers[9]   # Output
])
```

```python
model_without_preprocessing.layers
```

```python
# Export as TensorFlow SavedModel
model_without_preprocessing.export("/tmp/inception_v3_adam_savedmodel")
```
